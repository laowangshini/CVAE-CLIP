````python
# 安装必要的库
# 如果您尚未安装这些库，请取消下面代码的注释并运行
# !pip install torch torchvision pandas numpy matplotlib

import os
import pandas as pd
import numpy as np
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

# 检查GPU是否可用
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")

# 定义数据集类
class CelebADataset(Dataset):
    def __init__(self, img_dir, attr_path, bbox_path, partition_path, transform=None, partition=0):
        """
        初始化数据集
        :param img_dir: 图像文件夹路径
        :param attr_path: 属性文件路径
        :param bbox_path: 边界框文件路径
        :param partition_path: 分区文件路径
        :param transform: 图像预处理
        :param partition: 使用的数据分区 (0: train, 1: val, 2: test)
        """
        self.img_dir = img_dir
        self.transform = transform

        # 读取属性文件，使用逗号作为分隔符
        attr_df = pd.read_csv(attr_path, sep=',', header=0)
        
        # 读取分区文件，使用逗号作为分隔符
        partition_df = pd.read_csv(partition_path, sep=',', header=0)
        
        # 合并属性文件和分区文件，基于'image_id'列
        attr_df = attr_df.merge(partition_df, on='image_id')
        
        # 根据指定的分区进行筛选
        self.attr_df = attr_df[attr_df['partition'] == partition]
        
        # 读取边界框文件，使用逗号作为分隔符
        bbox_df = pd.read_csv(bbox_path, sep=',', header=0)
        
        # 合并边界框信息
        self.attr_df = self.attr_df.merge(bbox_df, on='image_id')

    def __len__(self):
        return len(self.attr_df)

    def __getitem__(self, idx):
        # 获取图像文件名
        img_name = self.attr_df.iloc[idx, 0]
        img_path = os.path.join(self.img_dir, img_name)
        
        # 打开图像并转换为RGB
        image = Image.open(img_path).convert('RGB')
        
        # 获取属性标签 (将-1和1转换为0和1)
        attrs = self.attr_df.iloc[idx, 1:41].values  # 假设属性列在第2列到第41列
        attrs = (attrs + 1) // 2  # 将-1转为0，1保持1
        attrs = attrs.astype(np.float32)
        
        # 应用图像预处理
        if self.transform:
            image = self.transform(image)
        
        return image, attrs

# 定义图像预处理
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # 调整图像大小
    transforms.ToTensor(),        # 转换为张量
    transforms.Normalize([0.5], [0.5])  # 归一化到[-1, 1]
])

# 数据集路径设置（请根据实际路径修改）
# 数据集路径设置（请根据实际路径修改）
img_dir = '/root/autodl-tmp/celeba_datasets/img_align_celeba/img_align_celeba'  # 图像文件夹
attr_path = '/root/autodl-tmp/celeba_datasets/list_attr_celeba.txt'  # 属性文件
bbox_path = '/root/autodl-tmp/celeba_datasets/list_bbox_celeba.txt'  # 边界框文件
partition_path = '/root/autodl-tmp/celeba_datasets/list_eval_partition.txt'  # 分区文件

# 创建训练集和验证集
train_dataset = CelebADataset(img_dir, attr_path, bbox_path, partition_path, transform=transform, partition=0)
val_dataset = CelebADataset(img_dir, attr_path, bbox_path, partition_path, transform=transform, partition=1)

# 数据加载器
batch_size = 128
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4)

# 定义CVAE模型
class CVAE(nn.Module):
    def __init__(self, img_channels=3, img_size=64, latent_dim=128, cond_dim=40):
        super(CVAE, self).__init__()
        self.img_size = img_size
        self.latent_dim = latent_dim
        self.cond_dim = cond_dim

        # 编码器部分
        self.encoder = nn.Sequential(
            nn.Conv2d(img_channels + cond_dim, 64, kernel_size=4, stride=2, padding=1),  # 32x32
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),  # 16x16
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),  # 8x8
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1),  # 4x4
            nn.BatchNorm2d(512),
            nn.ReLU(),
            nn.Flatten()
        )
        self.fc_mu = nn.Linear(512*4*4, latent_dim)
        self.fc_logvar = nn.Linear(512*4*4, latent_dim)

        # 解码器部分
        self.decoder_input = nn.Linear(latent_dim + cond_dim, 512*4*4)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),  # 8x8
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),  # 16x16
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),   # 32x32
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.ConvTranspose2d(64, img_channels, kernel_size=4, stride=2, padding=1),  # 64x64
            nn.Tanh()
        )

    def encode(self, x, c):
        # 将条件标签c与图像x拼接
        c = c.view(c.size(0), self.cond_dim, 1, 1).repeat(1, 1, self.img_size, self.img_size)
        x = torch.cat([x, c], dim=1)
        x = self.encoder(x)
        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        # 重参数化技巧
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z, c):
        # 将条件标签c与潜在变量z拼接
        z = torch.cat([z, c], dim=1)
        x = self.decoder_input(z)
        x = x.view(-1, 512, 4, 4)
        x = self.decoder(x)
        return x

    def forward(self, x, c):
        mu, logvar = self.encode(x, c)
        z = self.reparameterize(mu, logvar)
        recon_x = self.decode(z, c)
        return recon_x, mu, logvar

# 初始化模型、损失函数和优化器
latent_dim = 128
cond_dim = 40
model = CVAE(img_channels=3, img_size=64, latent_dim=latent_dim, cond_dim=cond_dim).to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.MSELoss(reduction='sum')  # 重建损失

# 定义损失函数
def loss_function(recon_x, x, mu, logvar):
    # 重建损失
    recon_loss = criterion(recon_x, x)
    # KL散度
    KL = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + KL

# 训练模型
num_epochs = 20
model.train()
for epoch in range(1, num_epochs + 1):
    train_loss = 0
    for batch_idx, (data, attrs) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}")):
        data = data.to(device)
        attrs = attrs.to(device)
        
        optimizer.zero_grad()
        recon_batch, mu, logvar = model(data, attrs)
        loss = loss_function(recon_batch, data, mu, logvar)
        loss.backward()
        train_loss += loss.item()
        optimizer.step()
    
    avg_loss = train_loss / len(train_loader.dataset)
    print(f"Epoch {epoch}, 平均损失: {avg_loss:.4f}")
    
    # 每个epoch后在验证集上评估
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for data, attrs in val_loader:
            data = data.to(device)
            attrs = attrs.to(device)
            recon_batch, mu, logvar = model(data, attrs)
            loss = loss_function(recon_batch, data, mu, logvar)
            val_loss += loss.item()
    avg_val_loss = val_loss / len(val_loader.dataset)
    print(f"验证集平均损失: {avg_val_loss:.4f}")
    model.train()

# 保存模型
torch.save(model.state_dict(), 'cvae_celeba.pth')
print("模型已保存为 cvae_celeba.pth")

# 定义生成函数
def generate_images(model, attrs, device, num_images=16):
    """
    根据条件标签生成图像
    :param model: 训练好的CVAE模型
    :param attrs: 条件标签 (N x 40)
    :param device: 设备
    :param num_images: 生成图像数量
    :return: 生成的图像
    """
    model.eval()
    with torch.no_grad():
        z = torch.randn(num_images, latent_dim).to(device)
        attrs = attrs.to(device)
        generated = model.decode(z, attrs)
        generated = generated.cpu()
        return generated

# 示例：生成图像
import torchvision

# 从验证集中随机选择一个批次的条件标签
data_iter = iter(val_loader)
images, attrs = next(data_iter)
sample_attrs = attrs[:16]

# 生成图像
generated_images = generate_images(model, sample_attrs, device, num_images=16)

# 可视化生成的图像
def show_images(images, title="Generated Images"):
    images = images * 0.5 + 0.5  # 反归一化
    grid = torchvision.utils.make_grid(images, nrow=4)
    np_grid = grid.numpy()
    plt.figure(figsize=(8,8))
    plt.imshow(np.transpose(np_grid, (1, 2, 0)))
    plt.title(title)
    plt.axis('off')
    plt.show()

show_images(generated_images, title="生成的人脸图像")

# 示例：条件控制生成（例如，改变某个属性）
# 这里以第一个样本的属性为例，将“Smiling”属性设为1
modified_attrs = sample_attrs.clone()
modified_attrs[:, 31] = 1  # 假设第32个属性是“Smiling”

# 生成带有修改属性的图像
modified_generated = generate_images(model, modified_attrs, device, num_images=16)
show_images(modified_generated, title="修改‘Smiling’属性后生成的人脸图像")
````

我来为您详细分析在这个条件变分自编码器(CVAE)中，条件(C)是如何体现的。

1. 训练过程中的条件体现：

在 `encode` 方法中：
```python
def encode(self, x, c):
    # 将条件标签c与图像x拼接
    c = c.view(c.size(0), self.cond_dim, 1, 1).repeat(1, 1, self.img_size, self.img_size)
    x = torch.cat([x, c], dim=1)
    x = self.encoder(x)
    mu = self.fc_mu(x)
    logvar = self.fc_logvar(x)
    return mu, logvar
```
条件体现在这里：
- 将属性标签 `c`（40维）扩展到图像大小（64x64）
- 将属性标签与图像特征在通道维度上拼接
- 这意味着编码器在学习特征时，会同时考虑图像本身和图像的属性标签

2. 生成过程中的条件体现：

在 `decode` 方法中：
```python
def decode(self, z, c):
    # 将条件标签c与潜在变量z拼接
    z = torch.cat([z, c], dim=1)
    x = self.decoder_input(z)
    x = x.view(-1, 512, 4, 4)
    x = self.decoder(x)
    return x
```
条件体现在这里：
- 将随机生成的潜在向量 `z` 与属性标签 `c` 拼接
- 这保证了生成的图像不仅依赖随机性，还受到属性标签的约束

具体示例：
```python
# 生成带有特定属性的图像
modified_attrs = sample_attrs.clone()
modified_attrs[:, 31] = 1  # 将"Smiling"属性设为1
modified_generated = generate_images(model, modified_attrs, device, num_images=16)
```
这里通过修改属性标签，可以控制生成图像的特定属性（如微笑），这正是条件变分自编码器的核心优势。

关键点：
- 条件信息 `c` 在编码和解码过程中都起作用
- 通过拼接条件信息，模型学习了图像特征与属性之间的关系
- 生成过程可以通过改变条件标签来控制生成图像的特征

这种设计使得CVAE不仅是一个生成模型，还是一个可以通过条件控制生成过程的生成模型。

在您的训练过程中，条件(C)主要体现在以下几个地方：

1. 数据加载阶段：
```python
for batch_idx, (data, attrs) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}")):
    data = data.to(device)
    attrs = attrs.to(device)
```
每次迭代都传入了图像`data`和对应的属性标签`attrs`。

2. 模型前向传播：
```python
recon_batch, mu, logvar = model(data, attrs)
```
在模型的前向传播中，同时传入了图像和属性标签。

3. 损失函数计算：
```python
loss = loss_function(recon_batch, data, mu, logvar)
```
尽管损失函数本身没有直接使用属性标签，但在模型的`forward`方法中，属性标签`attrs`已经参与了重建过程。

关键的条件体现在模型的`forward`方法中：
```python
def forward(self, x, c):
    mu, logvar = self.encode(x, c)  # 编码时使用条件
    z = self.reparameterize(mu, logvar)
    recon_x = self.decode(z, c)  # 解码时也使用条件
    return recon_x, mu, logvar
```

这意味着在训练过程中，模型通过属性标签`c`学习了图像特征与属性之间的关系，使得重建过程和潜在空间的表示都受到属性标签的条件约束。

简而言之，训练过程中条件的体现是通过在每一步前向和反向传播中，将属性标签与图像数据一起输入模型，让模型学习如何利用这些条件信息。

````python
# 安装必要的库
# 如果您尚未安装这些库，请取消下面代码的注释并运行
# !pip install torch torchvision pandas numpy matplotlib

import os
import pandas as pd
import numpy as np
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

# 检查GPU是否可用
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"使用设备: {device}")

# 定义数据集类
class CelebADataset(Dataset):
    def __init__(self, img_dir, attr_path, bbox_path, partition_path, transform=None, partition=0):
        """
        初始化数据集
        :param img_dir: 图像文件夹路径
        :param attr_path: 属性文件路径
        :param bbox_path: 边界框文件路径
        :param partition_path: 分区文件路径
        :param transform: 图像预处理
        :param partition: 使用的数据分区 (0: train, 1: val, 2: test)
        """
        self.img_dir = img_dir
        self.transform = transform

        # 读取属性文件，使用逗号作为分隔符
        attr_df = pd.read_csv(attr_path, sep=',', header=0)
        
        # 读取分区文件，使用逗号作为分隔符
        partition_df = pd.read_csv(partition_path, sep=',', header=0)
        
        # 合并属性文件和分区文件，基于'image_id'列
        attr_df = attr_df.merge(partition_df, on='image_id')
        
        # 根据指定的分区进行筛选
        self.attr_df = attr_df[attr_df['partition'] == partition]
        
        # 读取边界框文件，使用逗号作为分隔符
        bbox_df = pd.read_csv(bbox_path, sep=',', header=0)
        
        # 合并边界框信息
        self.attr_df = self.attr_df.merge(bbox_df, on='image_id')

    def __len__(self):
        return len(self.attr_df)

    def __getitem__(self, idx):
        # 获取图像文件名
        img_name = self.attr_df.iloc[idx, 0]
        img_path = os.path.join(self.img_dir, img_name)
        
        # 打开图像并转换为RGB
        image = Image.open(img_path).convert('RGB')
        
        # 获取属性标签 (将-1和1转换为0和1)
        attrs = self.attr_df.iloc[idx, 1:41].values  # 假设属性列在第2列到第41列
        attrs = (attrs + 1) // 2  # 将-1转为0，1保持1
        attrs = attrs.astype(np.float32)
        
        # 应用图像预处理
        if self.transform:
            image = self.transform(image)
        
        return image, attrs

# 定义图像预处理
transform = transforms.Compose([
    transforms.Resize((64, 64)),  # 调整图像大小
    transforms.ToTensor(),        # 转换为张量
    transforms.Normalize([0.5], [0.5])  # 归一化到[-1, 1]
])

# 数据集路径设置（请根据实际路径修改）
# 数据集路径设置（请根据实际路径修改）
img_dir = '/root/autodl-tmp/celeba_datasets/img_align_celeba/img_align_celeba'  # 图像文件夹
attr_path = '/root/autodl-tmp/celeba_datasets/list_attr_celeba.txt'  # 属性文件
bbox_path = '/root/autodl-tmp/celeba_datasets/list_bbox_celeba.txt'  # 边界框文件
partition_path = '/root/autodl-tmp/celeba_datasets/list_eval_partition.txt'  # 分区文件

# 创建训练集和验证集
train_dataset = CelebADataset(img_dir, attr_path, bbox_path, partition_path, transform=transform, partition=0)
val_dataset = CelebADataset(img_dir, attr_path, bbox_path, partition_path, transform=transform, partition=1)

# 数据加载器
batch_size = 128
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=4)

# 定义CVAE模型
class CVAE(nn.Module):
    def __init__(self, img_channels=3, img_size=64, latent_dim=128, cond_dim=40):
        super(CVAE, self).__init__()
        self.img_size = img_size
        self.latent_dim = latent_dim
        self.cond_dim = cond_dim

        # 编码器部分（移除了批量归一化层）
        self.encoder = nn.Sequential(
            nn.Conv2d(img_channels + cond_dim, 64, kernel_size=4, stride=2, padding=1),  # 32x32
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),  # 16x16
            nn.ReLU(),
            nn.Conv2d(128, 256, kernel_size=4, stride=2, padding=1),  # 8x8
            nn.ReLU(),
            nn.Conv2d(256, 512, kernel_size=4, stride=2, padding=1),  # 4x4
            nn.ReLU(),
            nn.Flatten()
        )
        self.fc_mu = nn.Linear(512*4*4, latent_dim)
        self.fc_logvar = nn.Linear(512*4*4, latent_dim)

        # 解码器部分
        self.decoder_input = nn.Linear(latent_dim + cond_dim, 512*4*4)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),  # 8x8
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),  # 16x16
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),   # 32x32
            nn.ReLU(),
            nn.ConvTranspose2d(64, img_channels, kernel_size=4, stride=2, padding=1),  # 64x64
            nn.Sigmoid()
        )

    def encode(self, x, c):
        # 将条件标签c与图像x拼接
        c = c.view(c.size(0), self.cond_dim, 1, 1).repeat(1, 1, self.img_size, self.img_size)
        x = torch.cat([x, c], dim=1)
        x = self.encoder(x)
        mu = self.fc_mu(x)
        logvar = self.fc_logvar(x)
        return mu, logvar

    def reparameterize(self, mu, logvar):
        # 重参数化技巧
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)
        return mu + eps * std

    def decode(self, z, c):
        # 将条件标签c与潜在变量z拼接
        z = torch.cat([z, c], dim=1)
        x = self.decoder_input(z)
        x = x.view(-1, 512, 4, 4)
        x = self.decoder(x)
        return x

    def forward(self, x, c):
        mu, logvar = self.encode(x, c)
        z = self.reparameterize(mu, logvar)
        recon_x = self.decode(z, c)
        return recon_x, mu, logvar

# 初始化模型、损失函数和优化器
latent_dim = 128
cond_dim = 40
model = CVAE(img_channels=3, img_size=64, latent_dim=latent_dim, cond_dim=cond_dim).to(device)

# 修改优化器的学习率
optimizer = optim.Adam(model.parameters(), lr=1e-4)

criterion = nn.MSELoss(reduction='sum')  # 重建损失

# 修改损失函数
def loss_function(recon_x, x, mu, logvar):
    # 重建损失，使用二元交叉熵损失
    recon_loss = criterion(recon_x, x)
    # 添加数值稳定性处理
    logvar = torch.clamp(logvar, min=-10, max=10)
    # KL 散度
    KL = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())
    return recon_loss + KL

# 训练模型
num_epochs = 10
model.train()

# 在训练循环中，添加梯度裁剪
for epoch in range(1, num_epochs + 1):
    train_loss = 0
    for batch_idx, (data, attrs) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch}/{num_epochs}")):
        data = data.to(device)
        attrs = attrs.to(device)

        optimizer.zero_grad()
        recon_batch, mu, logvar = model(data, attrs)
        loss = loss_function(recon_batch, data, mu, logvar)
        loss.backward()
        # 添加梯度裁剪
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        train_loss += loss.item()
        optimizer.step()

    # 计算并打印平均损失
    avg_loss = train_loss / len(train_loader.dataset)
    print(f"Epoch {epoch}, 平均损失: {avg_loss:.4f}")
    
    # 每个epoch后在验证集上评估
    model.eval()
    val_loss = 0
    with torch.no_grad():
        for data, attrs in val_loader:
            data = data.to(device)
            attrs = attrs.to(device)
            recon_batch, mu, logvar = model(data, attrs)
            loss = loss_function(recon_batch, data, mu, logvar)
            val_loss += loss.item()
    avg_val_loss = val_loss / len(val_loader.dataset)
    print(f"验证集平均损失: {avg_val_loss:.4f}")
    model.train()

# 保存模型
torch.save(model.state_dict(), 'cvae_celeba.pth')
print("模型已保存为 cvae_celeba.pth")

# 定义生成函数
def generate_images(model, attrs, device, num_images=16):
    """
    根据条件标签生成图像
    :param model: 训练好的CVAE模型
    :param attrs: 条件标签 (N x 40)
    :param device: 设备
    :param num_images: 生成图像数量
    :return: 生成的图像
    """
    model.eval()
    with torch.no_grad():
        z = torch.randn(num_images, latent_dim).to(device)
        attrs = attrs.to(device)
        generated = model.decode(z, attrs)
        generated = generated.cpu()
        return generated

# 示例：生成图像
import torchvision

# 从验证集中随机选择一个批次的条件标签
data_iter = iter(val_loader)
images, attrs = next(data_iter)
sample_attrs = attrs[:16]

# 生成图像
generated_images = generate_images(model, sample_attrs, device, num_images=16)

# 可视化生成的图像
def show_images(images, title="Generated Images"):
    images = images * 0.5 + 0.5  # 反归一化
    grid = torchvision.utils.make_grid(images, nrow=4)
    np_grid = grid.numpy()
    plt.figure(figsize=(8,8))
    plt.imshow(np.transpose(np_grid, (1, 2, 0)))
    plt.title(title)
    plt.axis('off')
    plt.show()

show_images(generated_images, title="生成的人脸图像")

# 示例：条件控制生成（例如，改变某个属性）
# 这里以第一个样本的属性为例，将“Smiling”属性设为1
modified_attrs = sample_attrs.clone()
modified_attrs[:, 31] = 1  # 假设第32个属性是“Smiling”

# 生成带有修改属性的图像
modified_generated = generate_images(model, modified_attrs, device, num_images=16)
show_images(modified_generated, title="修改‘Smiling’属性后生成的人脸图像")
````
